```python
import numpy as np
import pymc3 as pm
import pandas as pd
import matplotlib.pyplot as mplt
```

```python
total_of_Rolls=70
total_of_appeared=19
with pm.Model() as Dice_Model:
    p_uniform=pm.Uniform('p_uniform',lower=0,upper=1)
    p_likelihood=pm.Binomial('p_likelihood',n=total_of_Rolls,p=p_uniform,observed=total_of_appeared)

    trace_plot=pm.sample(3000,tune=3000,chains=4,cores=4)
    summary=pm.summary(trace_plot)
    print(summary)
    pm.traceplot(trace_plot)
    mplt.show()
```

```python
total_of_Rolls=70
total_of_appeared=19
with pm.Model() as Dice_Model:
    p_Beta=pm.Beta('p_Beta',alpha=1,beta=1)
    p_of_likelihood=pm.Binomial('p_of_likelihood',n=total_of_Rolls,p=p_Beta,observed=total_of_appeared)  
    
    trace_Beta = pm.sample(1000, tune = 1000,chains = 2)
    
    B_summary = pm.summary(trace_Beta)
        
    pm.traceplot(trace_Beta)
    
    print(B_summary)
    
    mplt.show()
    
    
```
