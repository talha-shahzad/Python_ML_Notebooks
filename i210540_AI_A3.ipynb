```python
# Python3 program to find the next optimal move for a player 
player, opponent = 'x', 'o'

# This function returns true if there are moves 
# remaining on the board. It returns false if 
# there are no moves left to play. 
def isMovesLeft(board):
    for i in range(3):
        for j in range(3):
            if (board[i][j] == '_'):
                return True
    return False

# This is the evaluation function as discussed 
# in the previous article ( http://goo.gl/sJgv68 ) 
def evaluate(b):
    # Checking for Rows for X or O victory. 
	for row in range(3):	 
		if (b[row][0] == b[row][1] and b[row][1] == b[row][2]):		 
			if (b[row][0] == player):
				return 10
			elif (b[row][0] == opponent):
				return -10

	# Checking for Columns for X or O victory. 
	for col in range(3):
		if (b[0][col] == b[1][col] and b[1][col] == b[2][col]):
			if (b[0][col] == player):
				return 10
			elif (b[0][col] == opponent):
				return -10

	# Checking for Diagonals for X or O victory. 
	if (b[0][0] == b[1][1] and b[1][1] == b[2][2]):
		if (b[0][0] == player):
			return 10
		elif (b[0][0] == opponent):
			return -10

	if (b[0][2] == b[1][1] and b[1][1] == b[2][0]):
		if (b[0][2] == player):
			return 10
		elif (b[0][2] == opponent):
			return -10

	# Else if none of them have won then return 0 
	return 0
```

# Simple MinMax

```python

def minimax(board, depth, isMax,count):
    score = evaluate(board)
    count[0]+=1
    for i in board:
        print(i)
    print()

    if (score == 10):
        return score 

    if (score == -10):
        return score 

    if (isMovesLeft(board) == False):
        return 0
    
    if (isMax):
        best = -1000

        for i in range(3):
            for j in range(3):
                if (board[i][j]=='_'):
                    board[i][j] = player 
                    best = max( best, minimax(board, depth + 1, not isMax,count))
                    board[i][j] = '_'
        return best
    else:
        best = 1000
        for i in range(3):
            for j in range(3):
                if (board[i][j] == '_'):
                    board[i][j] = opponent
                    best = min(best, minimax(board, depth + 1, not isMax,count))
                    board[i][j] = '_'
        return best 

```

```python
def findBestMove(board):
    bestVal = -1000
    bestMove = (-1, -1)
    cn=0
    run=0
    for i in range(3):
        for j in range(3):
            if (board[i][j] == '_'):
                count=[0]
                run+=1
                print(f"Run {run}\n")
                board[i][j] = player
                moveVal = minimax(board, 0, False,count)
                cn+=count[0]
                board[i][j] = '_'
                if (moveVal > bestVal):
                    bestMove = (i, j)
                    bestVal = moveVal
    print(f"total Moves : {cn}",)
    print("The value of the best Move is :", bestVal) 
    print() 
    return bestMove 
```

```python
def heuristic(board, player):
    score = 0
    
    for i in range(3):
        
        if board[i].count(player) == 2 and board[i].count('_') == 1:
            score += 10
        if board[i].count(opponent) == 2 and board[i].count('_') == 1:
            score -= 10
        
        col = [board[j][i] for j in range(3)]
        if col.count(player) == 2 and col.count('_') == 1:
            score += 10
        if col.count(opponent) == 2 and col.count('_') == 1:
            score -= 10
    
    diag1 = [board[0][0],board[1][1],board[2][2]]
    diag2 = [board[2][0],board[1][1],board[0][2]]
    
    if diag1.count(player) == 2 and diag1.count('_') == 1:
        score += 10
    if diag1.count(opponent) == 2 and diag1.count('_') == 1:
        score -= 10
    
    if diag2.count(player) == 2 and diag2.count('_') == 1:
        score += 10
    if diag2.count(opponent) == 2 and diag2.count('_') == 1:
        score -= 10
    
    return score
```

```python
def ABPruningWithHeuristic(board, depth, isMax, alpha, beta, count, player):
    score = evaluate(board)
    

    if score == 10:
        return score
    if score == -10:
        return score
    if isMovesLeft(board) == False:
        return 0
    count[0]+=1
    for i in board:
        print(i)
    print()


    if isMax:
        best = float('-inf')
        for i in range(3):
            for j in range(3):
                if board[i][j] == '_':
                    board[i][j] = player
                    heuristic_val = heuristic(board, player)
                    best = max(best, heuristic_val + ABPruningWithHeuristic(board, depth + 1, not isMax, alpha, beta, count, player))
                    alpha = max(alpha, best)
                    board[i][j] = '_'
                    if beta <= alpha:
                        break
        return best
    else:
        best = float('inf')
        for i in range(3):
            for j in range(3):
                if board[i][j] == '_':
                    board[i][j] = 'O' if player == 'X' else 'X'
                    heuristic_val = heuristic(board, player)
                    best = min(best, heuristic_val + ABPruningWithHeuristic(board, depth + 1, not isMax, alpha, beta, count, player))
                    beta = min(beta, best)
                    board[i][j] = '_'
                    if beta <= alpha:
                        break
        return best
```

```python
def findBestMoveABHeuristic(board):
    bestVal = -1000
    bestMove = (-1, -1)
    alpha = -1000
    beta = 1000
    cn=0
    run = 0
    for i in range(3):
        for j in range(3):
            if board[i][j] == '_':
                ABcount = [0]
                run += 1
                print(f"Run {run}\n")
                
                board[i][j] = player
                moveVal = ABPruningWithHeuristic(board, 0, False, alpha, beta, ABcount,player)
                cn+=ABcount[0]
                
                board[i][j] = '_'
                if moveVal > bestVal:
                    bestMove = (i, j)
                    bestVal = moveVal
                    alpha = max(alpha, bestVal)
    print(f"total Moves : {cn}",)
    print("The value of the best Move is:", bestVal) 
    print() 
    return bestMove
```
